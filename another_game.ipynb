```python
import string
import random

words = ['NodBootcamp', 'functions', 'programmer', 'jazz', 'beekeeper', 'jackpot', 'rhubarb', 'witchcraft', 'quizzes', 'zigzagging', 'zombie', 'subway', 'transcript']

hang_man = ['''
        +---+
            |
            |
            |
           ===''','''
        +---+
        O   |
            |
            |
           ===''','''
        +---+
        O   |
        |   |
            |
           ===''','''       
        +---+
        O   |
       /|   |
            |
           ===''','''  
        +---+
        O   |
       /|\  |
            |
           ===''','''  
        +---+
        O   |
       /|\  |
       /    |
           ===''',''' 
        +---+
        O   |
       /|\  |
       / \  |
           ===''']

def welcome(): 

    username = input("Please enter your username ")
    while not username.isalpha():
        print("Please enter characters A-Z only.")
        
        username = input("Please enter your username ")
    
    print(hang_man[0])
    print("Welcome to the Hangman game! ")
    print(f"Your username is {username}")
  
    print("\n")
    
    
def get_secret_word(words):
    return random.choice(words).upper()


def convert_word_to_uscore(word):
    lst = ["_"] * len(word)
    return " ".join(lst)


def input_guess():
    alpha = [i for i in string.ascii_lowercase + string.ascii_uppercase]
    guess = input("\nPlease guess a letter: ")
    
    while guess not in alpha:
        print("Please enter characters A-Z only. ")
        guess = str(input("Please guess a letter"))
        
    return guess.upper()


def check_guess(guess, word):
    if guess in word:
        return True
    else:
        return False
    
    
def update_word(guess, secret_word, good_guesses):
    good_guesses.append(guess)
    
    hidden_word = []
    for i in secret_word:
        if i in good_guesses:
            hidden_word.append(i)
        else:
            hidden_word.append("_")

    return " ".join(hidden_word)

    
def update_attempts_left(guess):
    bad_guesses.append(guess[0])
    return bad_guesses

```

```python
welcome()
print('\n')
secret_word = get_secret_word(words)

hidden_word = convert_word_to_uscore(secret_word)

print('\n')
print(hidden_word)

attempts = len(hang_man)
good_guesses = []

while attempts > 0:
    guess = input_guess()
    check = check_guess(guess, secret_word)
    
    if check == True:
        hidden_word = update_word(guess, secret_word, good_guesses)
        print('\n')
        print(hidden_word)
        if '_' not in hidden_word:
            print('\nYeahhhhh, you won!!')
            break
    else:
        attempts -= 1
        print(hang_man[-attempts])
        print("Attempts left: ", attempts -1)
```
